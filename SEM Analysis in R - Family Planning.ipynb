Next we are going to specify the model that we want to use to analyze these things according to the SEM method. We are going to save this model in the model variable.

The model is a 3-column matrix. You can see ncol=3 on the end of it, which specifies the the number of columns. byrow=TRUE on the end specifies that these are rows and not columns. If that would be FALSE instead of TRUE, then it would take the first 1/3 of the elements (because of 3 columens) and that would make 1 column.

Inside the matrix, we are specifxing in the vector (c()) how the connections look like. They are grouped together with three elements. The first specifies the connection. The second specifies the name of the connection and the third specifies the value of the connection. Either second or third can be unidentified, by using a NA. The ones that have the name specified will be the ones that will have the path calculated.

```r
model<-matrix(c(
"intention -> Intention.FutureChild", "i01", NA, 
"intention -> Intention.5YearChild", "i02", NA, 
"intention -> Intention.HandleChild", "i03", NA, 
"outcome -> Outcome.LessTime", "o01", NA, 
"outcome -> Outcome.Happier", "o02", NA, 
"outcome -> Outcome.LessBored", "o03", NA, 
"outcome -> Outcome.Relationship", "o04", NA, 
"outcome -> Outcome.Career", "o05", NA, 
"outcome -> Outcome.MoreCautious", "o06", NA, 
"outcome -> Outcome.FinanciallyInsecure", "o07", NA, 
"outcome -> Outcome.HopeInHumanity", "o08", NA, 
"outcome -> Outcome.LessSleep", "o09", NA, 
"outcome -> Outcome.SexDrive", "o10", NA, 
"outcome -> Outcome.Health", "o11", NA, 
"outcome -> Outcome.Meaning", "o12", NA, 
"outcome -> Outcome.SocialLife", "o13", NA, 
"controlbelief -> ControlBelief.FlexibleJob", "b01", NA, 
"controlbelief -> ControlBelief.FinancalSecurity", "b02", NA, 
"controlbelief -> ControlBelief.LivingArrangment", "b03", NA, 
"controlbelief -> ControlBelief.Marriage", "b04", NA, 
"controlbelief -> ControlBelief.StablePartner", "b05", NA, 
"controlbelief -> ControlBelief.AreaForChild", "b06", NA, 
"controlbelief -> ControlBelief.SucessfulCareer", "b07", NA, 
"controlbelief -> ControlBelief.HelpfulRelatives", "b08", NA, 
"current -> Current.FlexibleJob", "c01", NA, 
"current -> Current.Income", "c02", NA, 
"current -> Current.SupportivePartner", "c03", NA, 
"current -> Current.HelpfulFamily", "c04", NA, 
"current -> Current.SuitableArea", "c05", NA, 
"current -> Current.SuitableLivingArrangment", "c06", NA, 
"percivedbehcon -> PerBehaviourControl.WorkingSituation", "p01", NA, 
"percivedbehcon -> PerBehaviourControl.Income", "p02", NA, 
"percivedbehcon -> PerBehaviourControl.SupportivePartner", "p03", NA, 
"percivedbehcon -> PerBehaviourControl.SuitableArea", "p04", NA, 
"percivedbehcon -> PerBehaviourControl.SuitableLivingArragment", "p05", NA, 
"normativebel -> NormativeBelief.Family", "n01", NA, 
"normativebel -> NormativeBelief.Friends", "n02", NA, 
"normativebel -> NormativeBelief.Colleagues", "n03", NA, 
"normativebel -> NormativeBelief.Society", "n04", NA, 
"subjectivenor -> SubjectiveNorm.Family", "s01", NA, 
"subjectivenor -> SubjectiveNorm.Friends", "s02", NA, 
"subjectivenor -> SubjectiveNorm.Colleagues", "s03", NA, 
"subjectivenor -> SubjectiveNorm.Society", "s04", NA, 
"green -> normativebel", "g01", NA, 
"green -> subjectivenor", "g02", NA, 
"red -> controlbelief", "r01", NA, 
"red -> percivedbehcon", "r02", NA, 
"red -> current", "r03", NA, 
"purple -> outcome", "v01", NA, 
"intent <-> green", "f01", NA, 
"intent <-> red", "f02", NA, 
"intent <-> purple", "f03", NA, 
"intent -> intention", "r04", NA, 
"red <-> purple", "f04", NA, 
"green <-> purple", "f05", NA, 
"green <-> red", "f06", NA),
ncol=3, byrow=T)
```

Next I am actually calculating the model with method sem(). I am saving this into model.sem variable.

The first argument is the model, that I specified in the last step. The second argument in the covariance table. The N argument (third) is the number of observations that we have, which is number of people that solved survey in our case.

```r
model.sem <- sem(model, covDataFrame, N = 50)
```

Next, I am avreaging the variables in the same group. I am doing that by summing the vectors that present the variables that go to the same groups and then dividing by the number of vectors used.

If you are not comfortable with the linear algebra, then just imagine that for each row, I am checking the columns in questions, summing these numbers and dividing by the number of columns I checked.

```r
data3$Intention <- (data3$Intention.FutureChild + data3$Intention.5YearChild + data3$Intention.HandleChild)/3
data3$Outcome <- (data3$Outcome.LessTime + data3$Outcome.Happier + data3$Outcome.LessBored + data3$Outcome.Relationship + 
                 data3$Outcome.Career + data3$Outcome.MoreCautious + data3$Outcome.FinanciallyInsecure + 
                 data3$Outcome.HopeInHumanity + data3$Outcome.LessSleep + data3$Outcome.SexDrive + data3$Outcome.Health +
                 data3$Outcome.Meaning + data3$Outcome.SocialLife)/13
data3$ControlBelief <- (data3$ControlBelief.FlexibleJob + data3$ControlBelief.FinancalSecurity + 
                        data3$ControlBelief.LivingArrangment + data3$ControlBelief.Marriage + 
                        data3$ControlBelief.StablePartner + data3$ControlBelief.AreaForChild + 
                        data3$ControlBelief.SucessfulCareer + data3$ControlBelief.HelpfulRelatives)/8
data3$Current <- (data3$Current.FlexibleJob + data3$Current.Income + data3$Current.SupportivePartner + 
                 data3$Current.HelpfulFamily + data3$Current.SuitableArea + data3$Current.SuitableLivingArrangment)/6
data3$PerBehaviourControl <- (data3$PerBehaviourControl.WorkingSituation + data3$PerBehaviourControl.Income + 
                             data3$PerBehaviourControl.SupportivePartner + data3$PerBehaviourControl.SuitableArea + 
                             data3$PerBehaviourControl.SuitableLivingArragment)/5
data3$NormativeBelief <- (data3$NormativeBelief.Family + data3$NormativeBelief.Friends + data3$NormativeBelief.Colleagues + 
                         data3$NormativeBelief.Society)/4
data3$SubjectiveNorm <- (data3$SubjectiveNorm.Family + data3$SubjectiveNorm.Friends + data3$SubjectiveNorm.Colleagues + 
                        data3$SubjectiveNorm.Society)/4
```
