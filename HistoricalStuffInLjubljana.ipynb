# Visualizing Historical Stuff in Ljubljana (Testing Folium)

```python
from statistics import mean
import requests
import folium
import json
```

```python
overpass_url = "http://overpass-api.de/api/interpreter"
```

```python
overpass_query = """
[out:json];
area[name = "Ljubljana"];
(
  node["historic"](area);
  way["historic"](area);
  relation["historic"](area);
);
out center;
"""
```

```python
response = requests.get(overpass_url, 
                        params={'data': overpass_query})
data = response.json()
```

```python
coords = []
for element in data['elements']:
    if element['type'] == 'node':
        lat = element['lon']
        lon = element['lat']
        if "name" in element["tags"]:
            name = element["tags"]["name"]
        elif "note" in element["tags"]:
            name = element["tags"]["note"]
        if name:
            coords.append((lon, lat, name))
```

```python
m = folium.Map(location=[mean([a for a, b, c in coords]), mean([b for a, b, c in coords])], zoom_start=11)
for c in coords:
    folium.Marker([c[0], c[1]], popup=c[2], tooltip=c[2]).add_to(m)
```
