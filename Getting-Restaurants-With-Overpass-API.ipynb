# Getting Restaurants with Overpass API

```python
import requests
import json
```

```python
lat = 45.4774376
lon = 13.4998244
rad = 50000
country = 3600214885
```

```python
def createQuery(lat, lon, rad, country):
    query = ''
    query += '<osm-script output="json" output-config="" timeout="25">'
    query += '<id-query type="area" ref="' + str(country) + '" into="searchArea"/>'
    query +='<query into="_" type="node">'
    query +='<has-kv k="amenity" modv="" v="restaurant"/>'
    query +='<area-query from="searchArea"/>'
    query +='<around lat="' + str(lat) + '" lon="' + str(lon) + '" radius="' + str(rad) + '"/>'
    query +='</query>'
    query +='<print e="" from="_" geometry="skeleton" ids="yes" limit="" mode="body" n="" order="id" s="" w=""/>'
    query +='<recurse from="_" into="_" type="down"/>'
    query +='</osm-script>'
    return query
```

```python
query = createQuery(lat, lon, rad, country)
```

```python
data = requests.get('http://overpass-api.de/api/interpreter?data=' + query)
```

```python
data = data.text
```

```python
data = json.loads(data)
```

```python
all_resturants_names = []
for restaurant in data["elements"]:
    if "tags" in restaurant:
        tags = restaurant["tags"]
        if "name" in tags:
            all_resturants_names.append(tags["name"])
```
